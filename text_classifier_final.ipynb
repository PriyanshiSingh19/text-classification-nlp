```python
import os
import shutil
import sys
import subprocess

def run_cmd(cmd):
    print(f"\n>>> Running: {cmd}")
    result = subprocess.run(cmd, shell=True, text=True, capture_output=True)
    print(result.stdout)
    print(result.stderr)

run_cmd("pip uninstall -y tensorflow tensorflow-hub keras tf_keras")

site_packages = os.path.join(sys.prefix, "Lib", "site-packages")
folders_to_remove = ["tensorflow", "tensorflow_hub", "keras", "tf_keras"]

for folder in os.listdir(site_packages):
    for target in folders_to_remove:
        if folder.lower().startswith(target):
            path = os.path.join(site_packages, folder)
            try:
                shutil.rmtree(path)
                print(f" Removed leftover: {path}")
            except Exception as e:
                print(f" Could not remove {path}: {e}")

run_cmd("pip install tensorflow==2.9.0 tensorflow-hub==0.12.0 keras==2.9.0")

print("\n Reset complete! Now restart Jupyter Kernel/VS Code and test with:")
print("import tensorflow as tf; import tensorflow_hub as hub; print(tf.__version__, hub.__version__)")
```

```python
import tensorflow as tf
import tensorflow_hub as hub
from tensorflow.keras import layers, Sequential
import tensorflow_datasets as tfds
train_data, test_data = tfds.load(
    "imdb_reviews",
    split=['train', 'test'],
    as_supervised=True
)
BUFFER_SIZE = 10000
BATCH_SIZE = 32

train_data = train_data.shuffle(BUFFER_SIZE).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
test_data = test_data.batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
embedding_url = "https://tfhub.dev/google/nnlm-en-dim50/2"
hub_layer = hub.KerasLayer(
    embedding_url,
    input_shape=[],
    dtype=tf.string,
    trainable=True
)
model = Sequential([
    hub_layer,
    layers.Dense(16, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.summary()
history = model.fit(
    train_data,
    epochs=5,
    validation_data=test_data
)
```

```python
import tensorflow as tf
import tensorflow_hub as hub
import tensorflow_datasets as tfds
train_data, val_data, test_data = tfds.load(
    name="imdb_reviews",
    split=('train[:80%]', 'train[80%:]', 'test'),
    as_supervised=True
)
def format_example(text, label):
    return tf.squeeze(text), label   # remove extra dim

train_data = train_data.map(format_example)
val_data = val_data.map(format_example)
model = tf.keras.Sequential([
    hub.KerasLayer("https://tfhub.dev/google/nnlm-en-dim50/2",
                   input_shape=[],
                   dtype=tf.string,
                   trainable=True),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(1)
])
model.compile(
    optimizer='adam',
    loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
    metrics=['accuracy']
)
history = model.fit(
    train_data.shuffle(10000).batch(100),
    validation_data=val_data.batch(100),
    epochs=10,
    verbose=1
)
```

```python
results = model.evaluate(test_data.batch(100), verbose=2)
for name, value in zip(model.metrics_names, results):
    print("%s: %.3f" % (name, value))
```
